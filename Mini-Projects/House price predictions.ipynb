```python
import numpy as np                  # Numerical calculations
import pandas as pd                 # Dataset handling
import matplotlib.pyplot as plt     # Visualization
```

```python
data = {
    "Area":[800,900,1000,1100,1200,1300,1400,1500,1600,1700,
            1800,1900,2000,2100,2200,2300,2400,2500,2600,2700],
    
    "Price":[150000,170000,180000,200000,220000,
             235000,250000,265000,280000,300000,
             315000,330000,350000,370000,390000,
             405000,420000,440000,455000,480000]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
print(df.isnull().sum())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Area"],df["Price"],color='blue')
plt.title("House Area vs Price")
plt.xlabel("Area (Square Feet)")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```

```python
X = df[['Area']]     # Independent Variable
y = df['Price']   
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=39
)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import (
    mean_squared_error,
    mean_absolute_error,
    r2_score
)
```

```python
accuracy=r2_score(y_test,y_pred)
```

```python
accuracy
```

```python
mse=mean_squared_error(y_test,y_pred)
```

```python
mse
```

```python
plt.figure(figsize=(8,5))

plt.scatter(X,y,color='blue',label="Actual Data")

plt.plot(X,
         model.predict(X),
         color='red',
         linewidth=3,
         label="Best Fit Line")

plt.xlabel("Area")

plt.ylabel("Price")

plt.title("Best Fitted Regression Line")

plt.legend()

plt.grid(True)

plt.show()
```
