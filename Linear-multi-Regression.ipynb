```python
X = df.drop('Sales', axis=1)
Y = df['Sales']
X
```

```python
Y
```

```python
x_train, x_test , y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42, shuffle=True)

```

```python
x_train
```

```python
x_test
```

```python
y_test
```

```python
y_train
```

```python
model = LinearRegression()

model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)

# [163.3,	31.6,	52.9], prediction:array([17.0347724])

y_pred 
```

```python
mae = mean_absolute_error(y_test, y_pred)
print('Mean absolute error',mae)
mse = mean_squared_error(y_test, y_pred)

print('Mean squard error: ',mse)

r2 = r2_score(y_test, y_pred)

print('Accuracy (R2_Score):', r2)
```

```python
results = pd.DataFrame({
    'Actual': y_test.values,
    'Predicted': y_pred
})

print(results.head(10))
```

```python
plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred, alpha=0.5, color = 'blue', label= 'Sales Predictions')

# Add the best-fit perfect regression line (y = x)

prefect_fit = np.linspace(min(y_test), max(y_test), 100)
plt.plot(prefect_fit, prefect_fit, color = 'red', linestyle= '--', linewidth = 2, label='Prefect Fit(y=x)')
plt.xlabel('Actual Sales')
plt.ylabel('Predicted Sales')
plt.title('Actual vs Predicted Sales')
plt.grid()
plt.legend()
plt.show()
```
